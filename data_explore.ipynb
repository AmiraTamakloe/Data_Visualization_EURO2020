# 1st:'Match information'

```python
import pandas as pd

# Sheet: 'Match information'
df_match_info = pd.read_excel('EURO_2020_DATA.xlsx', sheet_name = 'Match information')

# Displaying the first five rows of the DataFrame
print(df_match_info.head())
```

```python
# drop unnecessary columns
df_match_info.drop(['MatchDay', 'Session', 'MatchMinute', 'InjuryTime', 'NumberOfPhases', 'Phase', 'MatchStatus', 'StadiumID', 'RefereeWebName', 'AssistantRefereeWebName', 'Humidity', 'Temperature', 'WindSpeed'], axis=1, inplace=True)

# Displaying the first five rows of the DataFrame
print(df_match_info.head())
```

```python
# Total 51 matches in EURO_2020
df_match_info
```

# 2nd: from Match info -> Match Stats

```python
# Sheet: 'Match Stats'
df_match_stats = pd.read_excel('EURO_2020_DATA.xlsx', sheet_name = 'Match Stats')

# Displaying the first five rows of the DataFrame
print(df_match_stats.head())
```

```python
# drop unnecessary columns
df_match_stats.drop(['IsHomeTeam', 'IsAwayTeam'], axis = 1, inplace=True)
# Displaying the first five rows of the DataFrame
print(df_match_stats.head())
```

```python
# Extract unique values in the 'StatsName' column
unique_stats_names = df_match_stats['StatsName'].unique()
print("Unique Stats Names:")
print(unique_stats_names)
```

```python
# Count the number of unique values in the 'StatsName' column
num_unique_stats= df_match_stats['StatsName'].nunique()
print("Number of unique StatsName values:")
print(num_unique_stats)
```

```python
# Count the occurrences of each 'StatsName'
stats_name_counts = df_match_stats['StatsName'].value_counts()
print("Counts of each StatsName:")
print(stats_name_counts)
```

# 3rd: Compared 'Match Event' and 'Match Stats' 

```python
# Sheet: 'Match Stats'
df_match_events = pd.read_excel('EURO_2020_DATA.xlsx', sheet_name = 'Match events')

# Displaying the first five rows of the DataFrame
print(df_match_events.head())
```

```python
# drop unnecessary columns
df_match_events.drop(['Minute', 'Phase', 'InjuryMinute','TeamFromID', 'TeamToID', 'PlayerFromID', 'PlayerToID','MatchEventAttribute'],axis = 1, inplace=True)
# Displaying the first five rows of the DataFrame
print(df_match_events.head())
```

```python
df_match_events
```

```python
# Extract unique values in the 'Event' column
unique_event = df_match_events['Event'].unique()
print("Unique_Event:")
print(unique_event)
```

```python
# Count the number of unique values in the 'Event' column
num_unique_event= df_match_events['Event'].nunique()
print("Number of unique Event values:")
print(num_unique_event)
```

```python
# Count the occurrences of each 'StatsName'
event_name_counts = df_match_events['Event'].value_counts()
print("Counts of each Event:")
print(event_name_counts)
```

```python
# The results of comparison: its better to choose 'Match Event' (38) than 'Match Stats' (193)
```

# Create a New dataframe: Combine 'df_match_info' and 'df_match_events'

```python
df = pd.merge(df_match_info,df_match_events, on=['MatchID', 'HomeTeamName', 'AwayTeamName'], how='inner')
print(df.head())
```
